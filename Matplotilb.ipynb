```python
# 범례 legend

data_dict = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}

plt.plot('data_x','data_y', data=data_dict, label = 'Price ($)')
plt.xlabel('X축', fontdict={'color':'b','size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r','size':14})
plt. legend(loc='lower right')
plt.show()

# 그래프 영역에 범례를 나타내려면 plot 함수에 label 문자열을 지정하고 legend() 함수를 호출한다.
```

```python
data_1 = {'data_x':[1,2,3,4,5],'data_y':[2,3,5,10,8]}
data_2 = {'data_x':[1,2,3,4,5],'data_y':[3,5,9,7,7]}

plt.plot('data_x','data_y', data=data_1, label = 'Price ($)')
plt.plot('data_x','data_y', data=data_2, label = 'Demand (#)')

plt.xlabel('X축', fontdict={'color':'b','size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r','size':14})
plt. legend(loc=0, ncol=2)
plt.show()
```

```python
# 타이틀 제목 설정

a = np.arange(0,2,0.2)

plt.plot(a,a,'r--')
plt.plot(a,a**2,'bo', markersize=20)
plt.plot(a , a**3, 'g-.', linewidth=10)

plt.grid(True, axis='y',color='blue', alpha=0.1)
plt.title('그래프 제목')

plt.show()
```

```python
# 막대 그래프 
x = np.arange(3)
years = ['2019','2020','2021']
values = [100,400,900]

plt.bar(x,values)
plt.xticks(x, years)

plt.show()
```

```python
x = np.arange(3)
years = ['2019','2020','2021']
values = [100,400,900]

plt.bar(x,values, width=0.6, align='edge', color='green')
plt.xticks(x, years)

plt.show()
```

```python
x = np.arange(3)
years = ['2019','2020','2021']
values = [100,400,900]

plt.barh(x,values, color='green')
plt.yticks(x, years)

plt.show()
```

```python
x = np.arange(3)
years = ['2019','2020','2021']
values = [100,400,900]

plt.barh(x,values, color='green', height=0.6, align='edge', edgecolor="gray")
plt.yticks(x, years)

plt.show()
```

```python
import numpy as np

np.random.seed(1)
n = 50

x = np.random.rand(n)
y = np.random.rand(n)
area = (30*np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x,y, s=area, c=colors, alpha=0.4, cmap='Set1')
plt.colorbar()
plt.show()
```

```python
import numpy as np

np.random.seed(1)
n = 50

x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

area = (30*np.random.rand(n)) ** 2
colors = np.random.rand(n)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, s=area, c=colors, alpha=0.4, cmap='Set1')

plt.show()
```

```python
# 타이타닉 시각화
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_excel('titanic.xls')
raw_data.info()
```

```python
raw_data.describe()
```

```python
raw_data.head()
```

```python
raw_data['survived'].value_counts().plot.pie(autopct='%1.2f%%')  #데이터에서 suvived를 카운트한 값
# autopct = %1.2f%%   1은 숫자를 찍고 .2는 소숫점 2자리까지 앞뒤로 %붙이고 f뒤에 %은 퍼센트라고 찍어라

plt.show()
```

```python
sns.countplot('survived', data = raw_data)  #씨본을 사용해서 raw_data에서 survived값을 세서 plot으로 그려라
plt.show()
```

```python
raw_data['age'].hist(bins=20, grid=False)
plt.show()
```
